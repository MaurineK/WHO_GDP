### Has life expectancy increased over time in the six nations?

```python
plt.figure(figsize=(6, 5))
sns.lineplot(x="Year", y="Life expectancy at birth (years)", hue="Country", data=data)
plt.title('Life Expectancy Over Time by Country')
plt.xlabel('Year')
plt.ylabel('Life Expectancy at Birth (years)')
plt.legend(title='Country', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
plt.clf()
```

### Has GDP increased over time in the six nations?

```python
plt.figure(figsize=(6, 5))
sns.lineplot(x="Year", y="GDP", hue="Country", data=data)
plt.title("GDP overtime by country")
plt.xlabel("Year")
plt.ylabel("GDP")
plt.legend(title="Country", bbox_to_anchor=(1.05, 1), loc="upper left")
plt.show()
plt.clf()
```

### Is there a correlation between GDP and life expectancy of a country?

```python
from scipy.stats import pearsonr

correlation, p_value = pearsonr(data["GDP"], data["Life expectancy at birth (years)"])
print(f"Correlation: {correlation}")
print(f"P Value: {p_value}")
```

```python
pd.crosstab(data["GDP"], data["Life expectancy at birth (years)"]
```

```python
sns.lmplot(x="GDP", y="Life expectancy at birth (years)", hue="Country", data=data, fit_reg = True)
plt.show()
plt.clf()
```

```python
sns.scatterplot(x="GDP", y="Life expectancy at birth (years)", data=data)
plt.title("Relationship between GDP and life expectancy")
plt.show()
plt.clf()
```

#### What is the average life expectancy in these nations?

```python
data["Life expectancy at birth (years)"].groupby(data["Country"]).mean().sort_values(ascending=False)
```

```python
data.groupby(["Country"])["Life expectancy at birth (years)"].mean().sort_values()
```
